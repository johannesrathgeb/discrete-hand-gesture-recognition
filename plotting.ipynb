```python
import numpy as np
import matplotlib.pyplot as plt

# Example accuracy matrices for demonstration
accuracy_matrix_og = np.array([
    [83.5, 85.9, 86.7],  # 50 users, 10, 30, 50 reps
    [88.4, 90.1, 88.4],  # 150 users, 10, 30, 50 reps
    [89.6, 91.1, 92.5],  # 300 users, 10, 30, 50 reps
])

accuracy_matrix_custom = np.array([
    [34.0, 83.0, 84.7],  # 50 users, 10, 30, 50 reps
    [85.0, 88.1, 89.0],  # 150 users, 10, 30, 50 reps
    [87.9, 90.5, 91.1],  # 300 users, 10, 30, 50 reps
])

difference_matrix = accuracy_matrix_custom -accuracy_matrix_og
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_matrix_seaborn(matrix, title, labels, cmap='viridis'):
    plt.figure(figsize=(8, 6))  # Adjusted size for better visibility
    ax = sns.heatmap(matrix, annot=True, fmt=".1f", cmap=cmap, cbar_kws={'label': 'Accuracy (%) / Difference (%)'})
    ax.set_title(title)
    
    # Adjust x and y ticks to center
    ax.set_xticks(np.arange(len(labels['x'])) + 0.5)  # Center ticks
    ax.set_xticklabels(labels['x'])
    ax.set_yticks(np.arange(len(labels['y'])) + 0.5)  # Center ticks
    ax.set_yticklabels(labels['y'])
    
    # Additional adjustments
    ax.set_xlabel("# of Reps per Gesture")
    ax.set_ylabel("# of Subjects in Training Set")
    ax.tick_params(axis='x', rotation=0)  # Optional: Rotate x-axis labels if needed

    plt.show()

# Labels for axes
labels = {
    'x': ['10', '30', '50'],
    'y': ['50', '150', '300']
}
```

```python
plot_matrix_seaborn(accuracy_matrix_og, "Accuracy Matrix Eddy", labels)
plot_matrix_seaborn(accuracy_matrix_custom, "Accuracy Matrix LSTM", labels)
plot_matrix_seaborn(difference_matrix, "Accuracy Difference Matrix compared to Eddy", labels)
```
